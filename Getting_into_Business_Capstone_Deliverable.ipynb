#Table 3: Mean, Standard Deviation, Median, Minimum, Maximum

Below shows a table displaying the mean, standard deviation, median, minimum, and maximum statistics for each attribute. These statistics help give us important insights to help make a good investment decision.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fratzcan/usa-house-prices")

print("Path to dataset files:", path)
```

This code pulls the data from the Kaggle USA housing data file and sorts it into mean, standard deviation, median, minimum, and maximums.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(path + "/USA Housing Dataset.csv")

# Select numeric columns for statistical analysis
numeric_df = df.select_dtypes(include=["number"])

# Compute summary statistics and round to one decimal place
summary_stats = pd.DataFrame({
    "Mean": numeric_df.mean().round(1),
    "Standard Deviation": numeric_df.std().round(1),
    "Median": numeric_df.median().round(1),
    "Maximum": numeric_df.max().round(1),
    "Minimum": numeric_df.min().round(1)
})

# Select categorical columns and compute mode
categorical_df = df.select_dtypes(include=["object"])
mode_values = categorical_df.mode().iloc[0]  # Get the first mode for each column

# Convert mode values into a DataFrame with the same format as numeric stats
mode_df = pd.DataFrame({"Mode": mode_values})

# Combine numerical stats and mode values into a single table
combined_stats = pd.concat([summary_stats, mode_df], axis=0)

# Capitalize row labels
combined_stats.index = combined_stats.index.str.capitalize()

# Convert numeric values to strings and remove trailing zeros (e.g., 3.500000 → 3.5)
combined_stats = combined_stats.astype(str).applymap(lambda x: x.rstrip('0').rstrip('.') if '.' in x else x)

# Replace NaN values with empty strings
combined_stats.replace("nan", "", inplace=True)

# Style table: center align headers and values
styled_combined = combined_stats.style.set_table_styles([
    {"selector": "th", "props": [("text-align", "center"), ("font-weight", "bold")]},  # Center headers
    {"selector": "td", "props": [("text-align", "center")]}  # Center data
])

# Display the final single summary table
display(styled_combined)
```

#Price Vs Living Square Feet

This scatterplot demonstrates how housing price changes as housing square feet changes. My hypothesis is as the square footage of a home and home price are positively correlated.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv(path + "/USA Housing Dataset.csv")

# Filter data to relevant ranges
filtered_df = df[(df["price"] <= 1_500_000) & (df["sqft_living"] <= 7000)]

# Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(filtered_df["sqft_living"], filtered_df["price"], alpha=0.5, color="blue", label="Data Points")

# Trend line (Linear Regression)
m, b = np.polyfit(filtered_df["sqft_living"], filtered_df["price"], 1)  # Fit line y = mx + b
plt.plot(filtered_df["sqft_living"], m * filtered_df["sqft_living"] + b, color="red", label="Trend Line")

# Labels and title
plt.xlabel("Living Area (sqft)")
plt.ylabel("Home Price ($)")
plt.title("Relationship Between Living Area and Home Price (Filtered)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

This shows that square footage of a home and home price are positively correlated. This could help in investment decisions because it shows that homes with more square feet are, on average, considered more valuable.

#Price Vs Bathrooms

This histogram demonstrates how housing price changes as the number of bathrooms in a house changes. My hypothesis is that number of bathrooms in a home and the price of a home are positively correlated. In addition I rounded the number of bathrooms to the nearest whole number.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fratzcan/usa-house-prices")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
path = "/root/.cache/kagglehub/datasets/fratzcan/usa-house-prices/versions/1"
df = pd.read_csv(path + "/USA Housing Dataset.csv")

# Round bathrooms and filter (1-5 bathrooms, price <= $1.5M)
df['bathrooms'] = df['bathrooms'].round()
filtered_df = df[(df['bathrooms'] >= 1) &
                 (df['bathrooms'] <= 5) &
                 (df['price'] <= 1_500_000)]

# Calculate average prices
avg_prices = filtered_df.groupby('bathrooms')['price'].mean().reset_index()

# Create the plot
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_prices['bathrooms'], avg_prices['price'],
               color='#1f77b4', edgecolor='black', width=0.6)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'${height:,.0f}',
             ha='center', va='bottom', fontsize=10)

# Formatting
plt.title("Average Home Price by Number of Bathrooms\n(1-5 Bathrooms, Rounded to Whole Numbers)",
          fontsize=14, pad=15)
plt.xlabel("Number of Bathrooms", fontsize=12)
plt.ylabel("Average Price ($)", fontsize=12)
plt.xticks(range(1, 6))  # Only show 1-5 on x-axis
plt.gca().yaxis.set_major_formatter('${x:,.0f}')
plt.grid(axis='y', alpha=0.2)

# Add trend arrow
max_bath = avg_prices['bathrooms'].max()
plt.annotate('Price Increases with More Bathrooms',
             xy=(max_bath, avg_prices['price'].iloc[-1]),
             xytext=(max_bath-2, avg_prices['price'].iloc[-1]*0.7),
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12)

plt.tight_layout()
plt.show()
```

This histogram demonstrates how as the number of bathrooms in a house increases the price of the house generally increases. While homes with 4 bathrooms have a higher average price than homes with 5 bathrooms for the most part homes with more bathrooms cost more. This data can help with investment decisions because it shows that homes with more bathrooms, on average, are considered more valuable.

#Price Vs Bedrooms

This histogram demonstrates how housing price changes as the number of bedrooms in a house changes. My hypothesis is that number of bedrooms in a home and the price of a home are positively correlated.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
path = "/root/.cache/kagglehub/datasets/fratzcan/usa-house-prices/versions/1"
df = pd.read_csv(path + "/USA Housing Dataset.csv")

# Round bedrooms and filter (1-6 bedrooms, price <= $1.5M)
df['bedrooms'] = df['bedrooms'].round()
filtered_df = df[(df['bedrooms'] >= 1) &
                 (df['bedrooms'] <= 6) &  # Adjust max bedrooms as needed
                 (df['price'] <= 1_500_000)]

# Calculate average prices
avg_prices = filtered_df.groupby('bedrooms')['price'].mean().reset_index()

# Create the plot
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_prices['bedrooms'], avg_prices['price'],
               color='#2ca02c',  # Green color
               edgecolor='black',
               width=0.6)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'${height:,.0f}',
             ha='center', va='bottom', fontsize=10)

# Formatting
plt.title("Average Home Price by Number of Bedrooms", fontsize=14, pad=15)
plt.xlabel("Number of Bedrooms", fontsize=12)
plt.ylabel("Average Price ($)", fontsize=12)
plt.xticks(range(1, int(avg_prices['bedrooms'].max()) + 1))  # Dynamic x-ticks
plt.gca().yaxis.set_major_formatter('${x:,.0f}')
plt.grid(axis='y', alpha=0.2)

# Add trend line
x = avg_prices['bedrooms']
y = avg_prices['price']
plt.plot(x, y, 'r--', alpha=0.5)  # Red dashed trend line
plt.scatter(x, y, color='red', s=50)  # Red dots at each data point

plt.tight_layout()
plt.show()
```

This demonstrates that the number of bedrooms in a home and the price of the home are positively correlated. This data can help with investment decisions because it shows that homes with more bedrooms, on average, are considered more valuable.
