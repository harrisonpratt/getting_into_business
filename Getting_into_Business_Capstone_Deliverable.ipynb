#Table 3: Mean, Standard Deviation, Median, Minimum, Maximum

Below shows a table displaying the mean, standard deviation, median, minimum, and maximum statistics for each attribute. These statistics help give us important insights to help make a good investment decision.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fratzcan/usa-house-prices")

print("Path to dataset files:", path)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(path+"/USA Housing Dataset.csv")

# Select numeric columns for statistical analysis
numeric_df = df.select_dtypes(include=["number"])

# Compute summary statistics and round to one decimal place
summary_stats = pd.DataFrame({
    "Mean": numeric_df.mean().round(1),
    "Standard Deviation": numeric_df.std().round(1),
    "Median": numeric_df.median().round(1),
    "Maximum": numeric_df.max().round(1),
    "Minimum": numeric_df.min().round(1)
})

# Select categorical columns and compute mode
categorical_df = df.select_dtypes(include=["object"])
mode_values = categorical_df.mode().iloc[0]  # Get the first mode for each column

# Convert mode values into a DataFrame with the same format as numeric stats
mode_df = pd.DataFrame({"Mode": mode_values})

# Combine numerical stats and mode values into a single table
combined_stats = pd.concat([summary_stats, mode_df], axis=0)

# Capitalize row labels
combined_stats.index = combined_stats.index.str.capitalize()

# Convert numeric values to strings and remove trailing zeros (e.g., 3.500000 → 3.5)
combined_stats = combined_stats.astype(str).applymap(lambda x: x.rstrip('0').rstrip('.') if '.' in x else x)

# Style table: center align headers and values
styled_combined = combined_stats.style.set_table_styles([
    {"selector": "th", "props": [("text-align", "center"), ("font-weight", "bold")]},  # Center headers
    {"selector": "td", "props": [("text-align", "center")]}  # Center data
])

# Display the final single summary table
display(styled_combined)
```

#Price Vs Living Square Feet

This scatterplot demonstrates how housing price changes as housing square feet changes. As the square footage in a home increases the price of the house increases.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv(path + "/USA Housing Dataset.csv")

# Filter data to relevant ranges
filtered_df = df[(df["price"] <= 1_500_000) & (df["sqft_living"] <= 7000)]

# Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(filtered_df["sqft_living"], filtered_df["price"], alpha=0.5, color="blue", label="Data Points")

# Trend line (Linear Regression)
m, b = np.polyfit(filtered_df["sqft_living"], filtered_df["price"], 1)  # Fit line y = mx + b
plt.plot(filtered_df["sqft_living"], m * filtered_df["sqft_living"] + b, color="red", label="Trend Line")

# Labels and title
plt.xlabel("Living Area (sqft)")
plt.ylabel("Home Price ($)")
plt.title("Relationship Between Living Area and Home Price (Filtered)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
